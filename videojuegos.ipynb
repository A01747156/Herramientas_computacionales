```python
# df["Year"].plot.box()
for col in numeric_columns :
    df[col].plot.box()
    plt.show()
```

```python
for col in numeric_columns :
    df[col].plot.kde(title = col)
    plt.show()
```

```python
print(f"Medidas de tendencia central")
print(f'Suma:{df['Year'].sum()}')
print(f"Promedio: {df["Year"].mean()}")
print(f"Mediana: {df["Year"].median()}")
print(f"Moda: {df["Year"].mode()}") # No con datos numéricos
print(f"Suma: {df["Year"].sum()}")
print(f"Suma: {df["Year"].sum()}")
```

```python
# df[máscara]
df[df["Year"] <= 1990]
```

```python
#Todos los juegos pubicados por nintendo después del 2000
# / or
# & and
# ~ not


df[(df['Year'] > 2000) & (df['Publisher'] == 'Nintendo')]['Name']
```

```python
# Mostrar los nombres de todos los juegos del ps, n64 y saturn
#df['Platform'].unique() Para saber que consolas hay

df[(df['Platform'] == 'PS') |
(df['Platform'] == 'N64') |
(df['Platform'] == 'SAT')]['Name']

# df[df['Platform'].isin()] Hace lo mismo que la de arriba pero con una línea
```
